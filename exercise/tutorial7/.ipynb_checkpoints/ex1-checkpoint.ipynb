```python
import numpy as np
import matplotlib.pyplot as plt

def fx(vx):
    return 2*vx+4*vx**0.5+2
def least_square(vx,vy):
    vx_mean=sum(vx)/len(vx)
    vy_mean=sum(vy)/len(vy)
    numer=0
    denom=0
    for i in range(len(vx)):
            numer += (vx[i] - vx_mean) * (vy[i] - vy_mean)
            denom += (vx[i] - vy_mean) ** 2
    w = numer / denom
    b = vy_mean - (w * vx_mean)
    return w,b
```

```python
x=np.linspace(0,4,20)
w1=[]
b1=[]
for i in range(10000):
    noise=1*np.random.randn(20)
    y=fx(x)+noise
    w1.append(least_square(x,y)[0])
    b1.append(least_square(x,y)[1])
plt.plot(w1,b1,'o')
```
